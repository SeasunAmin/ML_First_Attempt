```python
#showing the data set in a heatmap

plt.figure(figsize=(20,10))
sns.heatmap(data_frame.isnull())
```

```python
#checking the parsentage of missing data

parsentage_of_missing_value = data_frame.isnull().sum()/data_frame.shape[0]*100
print(parsentage_of_missing_value)
```

```python
#checking columns whos have missing data more then 17 persent

value_missing_columns = parsentage_of_missing_value[parsentage_of_missing_value >17].keys()
print(value_missing_columns)
```

```python
#droping missing colums and creat a new dataset/frame

data_frame_2 = data_frame.drop(columns=value_missing_columns)
data_frame_2.shape
```

```python
#showing 2nd data frame after fist cleaning

plt.figure(figsize=(20,10))
sns.heatmap(data_frame_2.isnull())
```

```python
#using dropna function we are deleting all the columns/row who have missing value

data_frame_3 = data_frame_2.dropna()
data_frame_3.shape
```

```python
#showing data frame 3 in heatmap

plt.figure(figsize=(20,10))
sns.heatmap(data_frame_3.isnull())
```
