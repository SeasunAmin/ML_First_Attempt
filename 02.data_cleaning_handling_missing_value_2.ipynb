```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("F://ml_assets//train.csv")
data_frame.shape
```

```python
#showing all the column and row of the dataset

pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
data_frame.head(6)
```

```python
data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
data_frame_missing_value_persent = data_frame.isnull().sum()/data_frame.shape[0]*100
print(data_frame_missing_value_persent)
```

```python
missing_value_more_than_20_persent = data_frame_missing_value_persent[data_frame_missing_value_persent >20].keys()
missing_value_more_than_20_persent
```

```python
data_frame_2 = data_frame.drop(columns=missing_value_more_than_20_persent)
data_frame_2.shape
```

```python
#taking numeric data from dataset2

data_frame_3_numeric_data = data_frame_2.select_dtypes(include=['int64','float64'])
data_frame_3_numeric_data.head()
```

```python
#heat map 

plt.figure(figsize=(20,10))
sns.heatmap(data_frame_3_numeric_data.isnull())
```

```python
#checking null values 

data_frame_3_numeric_data[data_frame_3_numeric_data.isnull().any(axis=1)]
```

```python
#total null values
data_frame_3_numeric_data.isnull().sum()
```

```python
#nemieric data missing persent

missing_value_persent_in_numeric_data =  data_frame_3_numeric_data.isnull().sum() / data_frame_3_numeric_data.shape[0]*100
missing_value_persent_in_numeric_data
```

```python
numeric_missing_value_cloumn = missing_value_persent_in_numeric_data[missing_value_persent_in_numeric_data>0].keys()
numeric_missing_value_cloumn
```

```python

for i, var in enumerate(numeric_missing_value_cloumn):
    plt.subplot(2,2,i+1)
    sns.displot(data_frame_3_numeric_data[var],bins=10)
```
